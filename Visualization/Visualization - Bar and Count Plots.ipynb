## Bar Charts using Matplotlib/Seaborn
- Bar plots are an effective visualization technique when we have categorical data that consists of various categories or groups.
- Example we can use bar plots to view test loan requests categorized by gender. These plots allow us to easily see the difference between categories (gender) because the size of the bar varies with the magnitude of the represented variable (loan amount), and categories are clearly divided and color-coded.

- However, Bar plots work well when we are dealing with a few categories.Too many categories will make the bars cluttered and the plots become hard to understand or complicated.

Notice that bar plots look very similar to histograms. How do we distinguish between the two? What are the differences between the two?

A histogram represents the frequency distribution of continuous variables while a bar graph is a diagrammatic comparison of discrete variables. The histogram presents numerical data whereas the bar graph shows categorical data. The histogram is drawn in such a way that there is no gap between the bars.

### Seaborn BarPlot
- We create  bar plots using the barplot() command in Seaborn and takes two arguments.
- The first argument to the command is the column to be used for the groups.
- The second argument is the column to be used for comparison.

```python
sns.set_style("darkgrid")
```

```python
plt.title("Department Rating")
sns.set(rc={'figure.figsize':(10,6)})
sns.barplot(x="Department Name", y='Rating' , data=df, hue="Division Name" )
```

```python
plt.title("Division Name vs Rating")
plot1=sns.barplot(x="Division Name", y='Rating' , data=df , hue='Department Name')
```

```python
sns.set(rc={'figure.figsize':(7,8)})
plt.title('Division Name')
sns.barplot( x="Division Name", y = 'Rating', data=df ,order=["Initmates", "General Petite","General"]);
```

## Count Plot
- https://seaborn.pydata.org/generated/seaborn.countplot.html
- Show the counts of observations in each categorical bin using bars.

A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
df1=pd.read_csv('Food_Preference.csv')
```

```python
df1.head()
```

```python
df1["Dessert"].unique()
```

```python
df1['Dessert'].value_counts()
```

```python
sns.countplot(x="Juice", data=df1)
```

```python
sns.countplot(y="Dessert", data=df1)
```

```python
# research on seaborn on themes
sns.set(rc={'figure.figsize':(7,8)})
sns.countplot(x="Food", data=df1)
```

```python
sns.countplot(x="Dessert", data=df1, hue="Gender", palette="Set1")
```

```python
DessertOrder = df['Dessert'].value_counts().sort_values(ascending=False).index
```

```python
## plot a countplot horizontally
sns.set(rc={'figure.figsize':(6,6)})
plt.title('Dessert Stats')
sns.countplot(y="Dessert", data=df, palette="Set1", order=DessertOrder)
```
