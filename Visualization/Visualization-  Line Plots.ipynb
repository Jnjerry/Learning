# Line  Plot Using Matplotlib / Seaborn
A line plot displays data points connected by straight line segments, instead of showing them individually. This type of plot is useful for finding relationships between datasets 

## When to use a line Plot

- Line charts are great for:

1. Comparing lots of data all at once

2. Showing changes and trends over time

3. Displaying forecast data and uncertainty

5. Highlighting anomalies within and across data series

##  Matplotlib
- Use the plt.plot() function to map the x and y values

https://matplotlib.org/stable/tutorials/introductory/pyplot.html


```python
year = [1970, 1980, 1990, 2000, 2010, 2020]
pop_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
pop_B = [44.4, 55.6, 69.7, 87.1, 95.4, 100.5]


#Visualize the Data
plt.plot(year, pop_A, color='g')
plt.plot(year, pop_B, color='r')
plt.xlabel('Countries')
plt.ylabel('Population in million')
plt.title('Populations over the years')

```

## Use the dataset above to plot a line plot using matplotlib

```python
from matplotlib.pyplot import figure
figure(figsize=(20, 10), dpi=80)
plt.title("Stock Prices for IBM(2019-2020)")
plt.xlabel("Date")
plt.ylabel("Opening and Closing Stock")
plt.plot(df["Date"],df[['High',"Low"]])
```

### Line Plot - Seaborn
- The lineplot() function of Seaborn is what plots the line graph. The first parameter is the list or array of x-values and the second parameter is the array of y-values.
- https://seaborn.pydata.org/generated/seaborn.lineplot.html

```python
df.head()
```

```python
sns.lineplot(x=df['Date'],y=df["High"])
```

```python
figure(figsize=(20,10))
sns.lineplot(data=df , x="Date",  y="High")
plt.title("IBM High Prices")
```
